```python
import datetime
```

##### Función transforma un datetime en un string con valor igual al tiempo(time)

```python
def transformarTiempo(tiempo):
    st = datetime.datetime.strptime(str(tiempo), "%Y-%m-%d %H:%M:%S")
    return st.time()
```

##### Función para imprimir Tiempo, recibe variables en el siguiente orden: Hora Actual, LLegada del Cliente, Fin de Servicio, N° Cola (q) y Estado del Puesto de Servicio(PS)'

debug imprime el paso de un tiempo al tiempo actual, de esta forma se puede saber por dónde recorrió el programa

```python
def imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, debug):
    

    sta = transformarTiempo(ta)
    stl = transformarTiempo(tl)
    stfs = transformarTiempo(tfs)
    stcd = transformarTiempo(tcd)
    stct = transformarTiempo(tct)
    
    
    if PS==0:
        stfs = "        "
    if S==0:
        stcd = "        "
    if S==1:
        stct = "        "
        
    print(sta," | ",stl," | ",stfs," | ",stcd," | ",stct," | ",q," | ",PS," | ",S," | ",debug)
```

##### Función vector de deltas de tiempo por minuto'

```python
def timedelta(i):
    return datetime.timedelta(seconds = i)
```

##### Declaración de los tiempos iniciales y los deltas

```python
#Instancia de Tiempo Actual'

ta = datetime.datetime.combine(datetime.datetime.today(),datetime.time(10,30,0,0))

#Instancia del limite(HORA DE SERVICIO)'
limite = ta + timedelta(60*60*1.5)

tl = datetime.datetime.combine(datetime.datetime.today(),datetime.time(10,30,20,0))
tcd  = datetime.datetime.combine(datetime.datetime.today(),datetime.time(10,31,34,0))
tfs = limite
tct = tcd


#Condiciones iniciales propia de caso, delta Tiempo de llegada y delta Tiempo de Fin de serv.'
dl = [timedelta(65), timedelta(6), timedelta(2), timedelta(21), timedelta(42)]
dfs = [timedelta(5), timedelta(10)]
descanso = timedelta(60)
dcd = timedelta(30)

```

##### Declaración de las variables

```python
#Condiciones Iniciales'

q = 0
PS = 0
S = 1

#Contadores'
cl = 0
cfs = 0

print(limite)
```

##### Programa principal, un while que imprime toda la matriz.

```python
print("  Actual  |  LLegada   | Fin de S.  |  Descanso  |  Vuelta T. |  q  |  PS |  S  |debug")
print("--------------------------------------------------------------------------------------")

#Se imprime primero el estado inicial
imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tl")


while ta < limite:
        #Si el tiempo de llegada de cliente es menor al fin de servicio'
        if tl < tfs and tl <= tcd:
            ta = tl
            tl += dl[cl]
            cl+=1
            
            #Caso especial: Si el Puesto de Servicio es 0 entonces'
            # tiempo de fin de servicio iniciará a partir del tiempo actual'
            if PS==0:
                PS = 1
                tfs = ta + dfs[cfs]
                cfs+=1
            else:
                q+=1
            
            imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tl")
            
            #Si el tiempo de fin de servicio es menor que el de llegada'
        elif tfs < tl and tfs <= tcd and PS == 1:
            ta = tfs
           
            
            if q > 0:
                q-= 1
                tfs += dfs[cfs]
                cfs+=1
            else:
                PS = 0
                tfs = limite
        
            imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tfs")
            
        #Entrada al descanso'
        elif tcd < tl and tcd < tfs and tcd < limite:
            ta = tcd
            tfs += descanso
            tct = ta + descanso
            S = 0
            imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tcd")
            
            #Mientras no se llegue al tiempo de vuelta al servicio se imprimiran'
            #todos los clientes que irán llegando'
            while tl <= tct:
                ta = tl
                tl += dl[cl]
                cl +=1
                
                if PS == 0:
                    PS = 1
                    tfs = tct + dfs[cfs]
                    cfs+=1
                else:
                    q+= 1
                imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tl")
                
                if cl > 4:
                    cl = 0
                if cfs > 1:
                    cfs = 0 
                
            #Vuelta al servicio, se aplica el delta de 30'' para volver al descanso'         
            S= 1
            ta = tct
            tcd = ta + dcd 
            imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tct")
        
        #Si el tiempo de entrada al descanso es mayor que el limite pero menor que los otros tiempos'
        #Entonces directamente se sale del bucle y se pone fin al servicio'
        elif tcd >= limite and tcd < tl and tcd < tfs:
            break
                
                
        #SI el tiempo de llegada es igual al tiempo de fin de serv., entonces se imprimen'
        #en lineas separadas'
        elif tl == tfs and tl <= tcd:
            ta = tl
            tl += dl[cl]
            cl+=1
            q+=1
            
            imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tl")
            
            ta = tfs
            tfs += dfs[cfs]
            cfs+=1
            if q == 1:
                q = 0
            else:
                q-= 1
            imprimirTiempo(ta,tl,tfs,tcd, tct, q ,PS, S, "tfs")
            
            
        #Debug'
        else:
            print("error")
            break
        
        #Reinicio de contadores de las listas'
        if cl > 4:
            cl = 0
        if cfs > 1:
            cfs = 0

print("Clientes sin atención:",q)
    
```
